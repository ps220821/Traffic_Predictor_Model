```python
snelheid.replace(r'^\s*$', np.nan, regex=True)
snelheid = snelheid.dropna(axis=1, how="all")
snelheid.head()
```

```python
keep_columns = [
    "avg_intensity",
    "avg_speed",
    "weighted_avg_speed",
    "total_number_of_lanes",
    "road_classification_1",
    "driving_direction",
    "road_number_vild",
    "start_measurement_period",     # start_meetperiode
    "name_measuring_location",
    'total_number_of_lanes'
]

snelheid.columns = snelheid.columns.str.strip()   # schoonmaken

# Neem alleen kolommen die in je keep_columns lijst voorkomen
existing_columns = [c for c in keep_columns if c in snelheid.columns]
snelheid = snelheid[existing_columns]
snelheid.head(30)
```

```python
# Filter out rows where avg_intensity < 1 or avg_speed < 1
snelheid = snelheid[(snelheid['avg_intensity'] >= 1)
                    & (snelheid['avg_speed'] >= 1)]

snelheid = snelheid[snelheid["driving_direction"] != "NaN"]

snelheid.head()
```
